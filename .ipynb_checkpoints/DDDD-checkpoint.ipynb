# Stuart's DDDD Dataset
#### Imputed mean for null values

```python
import Configure_Dataset 
DDDD, y1, y2 = Configure_Dataset.configure('Competition1_raw_data.xlsx')
DDDD.head()
```

```python
DDDD.isnull().sum()
```

```python
import Impute
DDDD = Impute.impute_mean(DDDD)
```

```python
DDDD.isnull().sum()
```

```python
#importnumpy ndarray for use in sklearn
import Normalize
```

```python
#transorm a pandas dataframe using a z-score transformation to return a pandas dataframe

DDDD = Normalize.z_score(DDDD)
```

```python
DDDD.head()
```

```python
#input a pandas dataframe of competition one dataset and returns a transformed dataset based on values of skewness for each column.
#    For a single column: 
#        if the value of skew is between [-.5, .5], no transformation takes place.
#        if the value of skew is less than -0.5, perform an e^x transformation.
#        if the value of skew is greater than .5, perform a log transformation.
#        '''

import Transform
```

```python
DDDD = Transform.skew_transform_log(DDDD)
```

```python
DDDD.head()
```

```python
import Combine
DDDDdata = Combine.combine(DDDD, y1, y2)
DDDDdata.head()
```
