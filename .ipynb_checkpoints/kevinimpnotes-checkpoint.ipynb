## Filling in missing values of continuous variables

For the remaining variables, all but one are continuous fields (C2 is binary), with multiple missing values. Our strategy here will be to use more novel imputing strategies. We will impute missing values into all remaining fields, and for C2 we will round our imputed values to either 0 or 1 at the end.  

For this section, we will use the python library 'impyute', which has many methods of imputation in a dataset. First we will need to properly format our dataset. 
- (Documentation: https://impyute.readthedocs.io/en/master/user_guide/overview.html)


```python
import impyute

# impytute requires a numpy array, so let's format one based on our data.
# Rather than transforming textual data,
# we simply drop these columns, as they are only identification-type columns.

imp_df = df.drop(columns=['I1', 'I2', 'I3'])
imp_df = imp_df.values


#verify nulls exist
np.isnan(imp_df).any()
```

```python
# will use impyute.imputation class to try out different imputation strategies listed
imp_rand_df = impyute.imputation.cs.random(imp_df) #takes a random number from each column and insterts into null
imp_mean_df = impyute.imputation.cs.mean(imp_df) #takes mean of column and inserts it into null

#check to see if all NaN's are filled
np.isnan(imp_rand_df).any()
```

Our main strategy for imputation is MICE. Here is a brief explanation of this method: __EXPLANATION HERE:__

Multivariate Imputation by Chained Equations:

```python
# MICE imputation
imp_mice_df = impyute.imputation.cs.mice(imp_df)
np.isnan(imp_mice_df).any()
```

```python
imp_mice_df[0:5]
```

Need to add back the columns that we dropped earlier to our new clean dataframe.

```python
# Convert numpy array back to pandas dataframe
clean_df = pd.DataFrame(imp_mice_df)

# Adding back in column names
clean_df.columns = ['P(IPO)', 'P(H)', 'P(L)', 'P(1Day)', 'C1', 'C2', 'C3',
       'C4', 'C5', 'C6', 'C7', 'T1', 'T2', 'T3', 'T4', 'T5', 'S1', 'S2', 'S3']

# joining the textual characteristics to the new clean_df
clean_df = clean_df.join(df[['I1', 'I2', 'I3']])

#verify that all nulls are filled. 

clean_df

```

Lastly, since we have a boolean column, we need to fix any decimal values for C2. We'll simply round the answers:
- if C2 < .5, C2 = 0
- if C2 > .5, C2 = 1

```python
# rounding the values
clean_df = clean_df.round({'C2': 0})
clean_df.C2
```
