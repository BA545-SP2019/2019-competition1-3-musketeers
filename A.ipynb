# Stuart's A Dataset
#### Imputed mean for null values
#### Normalized using minmax for null values

```python
import Configure_Dataset 
A, y1, y2 = Configure_Dataset.configure('Competition1_raw_data.xlsx')
A.head()
```

```python
A.isnull().sum()
```

```python
import Impute
A = Impute.impute_mean(A)
```

```python
A.isnull().sum()
```

```python
import Normalize
```

```python
A = Normalize.minmax_scale(A)
```

```python
A
```

```python
import Combine
Adata = Combine.combine(A, y1, y2)
Adata.head()
```

```python
import pandas as pd
Adata.to_csv('./Datasets/A.csv')
```
