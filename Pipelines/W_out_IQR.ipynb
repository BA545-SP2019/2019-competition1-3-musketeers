```python
#Outlier Treatment
   
def OutliersUpperBand(df):    
    for i in df.describe().columns:
        Q1=df.describe().at['25%',i]
        Q3=df.describe().at['75%',i]
        IQR=Q3 - Q1
        LTV=Q1 - 1.5 * IQR
        UTV=Q3 + 1.5 * IQR
        x=np.array(df[i])
        p=[]
        for j in x:
            if j < LTV:
                p.append(df[i].mean())
            elif j > UTV:
                p.append(df[i].mean())
            else:
                p.append(j)
        df[i]=p
    return df
OutliersUpperBand(df[["C1","C4","C7","%_long_sentences","%_long_words","%_positive_words","%_negative_words","%_uncertain_words","C5'","C6'"]])
```
