```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel("Competition1_raw_data.xlsx", header=0)
```

```python
df = df.replace('-', np.NaN)
```

```python
df= df.dropna()
```

```python
df.drop([174, 187,499,620,246,221])
  
```

```python
df = df.drop(['I1','I2'], axis = 1)
df.reset_index(drop=True, inplace = True)
```

```python
df['p_mid'] = (df['P(H)'] + df['P(L)']) / 2
```

```python

    
df["Y1"] = np.where(df['P(IPO)'] < df['p_mid'], 1, 0)
df['Y2'] = np.where(df['P(IPO)'] < df['P(1Day)'], 1, 0)
```

```python
df["C3'"] = np.where(df['C3'] > 0, 1, 0)
df["C5'"] = df['C5'] / df['C6']
```

```python
df['fraction'] = (df['P(IPO)'] - df['p_mid']) / df['p_mid'] * 100
df["C6'"] = np.where(df["P(IPO)"] > df['p_mid'], df['fraction'], 0) 
```

```python
y1 = df['Y1']
y2 = df['Y2']
df = df.drop(labels = ['p_mid', 'I3', 'P(IPO)', 'P(H)', 'P(L)', 'P(1Day)', 'fraction', 'C3', 'C5', 'C6', 'T1', 'T2', 'T3', 'T4', 'T5', 'S1', 'S2', 'S3', 'Y1', 'Y2'], axis = 1)
```

```python

df = df.join([y1, y2])
```
